```python
import pandas as pd
import numpy as np
import math
```

```python
def bert_format(df):
    df = pd.DataFrame({'guid': df['guid'],
                        'text_a': df['text_a'],
                        'text_b': pd.Series("a", index=np.arange(len(df))),
                        'theme1': df['theme1'].astype(int)})
    return df
```

```python
def widen_labels(df):
    size = len(df)
    labels = np.zeros((size, 9))
    labels[np.arange(size), df['theme1']-1] = 1
    
    labels_df = pd.DataFrame(labels)
    labels_df.columns = np.arange(9)+1
    
    df_wide = pd.DataFrame({"guid": df['guid'],
                             "text_a": df['text_a'],
                             "text_b": df['text_b']})
    df_wide = pd.concat([df_wide, labels_df], axis = 1)
    return df_wide
```

```python
for i in range(5):
    path = "dataset_multiclass_wide/"+str(i)+"/"
    test_labeled = pd.read_csv(path+"test_labeled.tsv",
                              sep='\t', header=None)
    train = pd.read_csv("dataset_multiclass_wide/"+str(i)+"/train.tsv",
                              sep='\t', header=None)
    
    cols = ['guid', 'text_a', 'theme1', 'theme2']
    test_labeled.columns = cols
    train.columns = cols
    
    test_labeled = bert_format(test_labeled)
    train = bert_format(train)
    
    test_labeled_wide = widen_labels(test_labeled)
    train_wide = widen_labels(train)
    
    display(test_labeled_wide.head())
    display(train_wide.head())
    
    test_labeled_wide.to_csv(path+"dev.tsv", sep='\t', header=False, index=False)
    train_wide.to_csv(path+"train.tsv", sep='\t', header=False, index=False)
```
