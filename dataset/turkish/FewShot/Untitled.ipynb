```python
import pandas as pd 
from sklearn.model_selection import StratifiedKFold
import numpy as np
```

```python
data = pd.read_csv('turkish_dev-Copy1.csv', header=0)
data.head()
```

```python
data.columns = ['guid','text_a','theme1','theme2']
```

```python
k=2
skf = StratifiedKFold(n_splits=k, random_state=50, shuffle=True)
X = data['text_a']
y = data['theme1']
skf.get_n_splits(X,y)
```

```python
for train_index, test_index in skf.split(X,y):
    print("dev:", len(train_index), "few shot train:", len(test_index))
    dev, few_shot_train = data.loc[train_index], data.loc[test_index]
    break
```

```python
dev_100 = pd.read_csv('dev-Copy1.tsv', sep='\t', header=None)
dev_100.loc[dev.index]
```

```python
len(dev.index)
```

```python
dev = dev_100.loc[dev.index.union(few_shot_train.index[:11])]
train = dev_100.loc[few_shot_train.index[11:]]
print(len(dev),len(train))
```

```python
dev.to_csv('dev.tsv', sep='\t', header=False, index=False)
train.to_csv('train.tsv', sep='\t', header=False, index=False)
```
