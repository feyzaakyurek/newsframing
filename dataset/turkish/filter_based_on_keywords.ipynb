```python
import pandas as pd
import os
```

```python
data_dir = "/project/llamagrp/samarth/relevance/data/other_countries"
```

```python
file_names = ["Australia_1.xls",
              "Australia_2.xls",
              "Canada_all.xls",
              "UK_1.xls",
              "UK_2.xls",
              "UK_3.xls",
              "UK_4.xls"]
```

```python
list_of_df = []
for f in file_names:
    print(f)
    df = pd.read_excel(os.path.join(data_dir, f), header=1)\
            .iloc[:,[0,1,17]]
    df.columns = ["guid", "date", "headline"]
    display(df.head())
    list_of_df.append(df)
    
    
    
# for f in file_names[1:]:
#     df_temp = pd.read_excel(os.path.join(data_dir, f)).iloc[:,[0,17]]
#     df_temp.columns = ["guid", "headline"]
#     print(df.columns)
#     df = pd.concat([df, df_temp])
#     print(df.columns)
    
```

```python
def filter_keywords(df):
    search_for_1 = ["gun","firearm","nra", "2nd amendment","second amendment","AR15", 
                    "assault weapon", "rifle", "brady act", "brady bill","shooting"]
    search_for_2 = ["United States", "USA", "America", "Trump"]
    
    result_1 = df['headline'].str.contains('|'.join(search_for_1),case=False)
    result_2 = df['headline'].str.contains('|'.join(search_for_2),case=False)
    result = result_1 & result_2
    print("Length of df: ", len(df))
    print("Number of headlines including words related to gun violence: ", result_1.sum())
    print("Number of headlines including words related to the US: ", result_2.sum())
    print("Number of headlines that include at least one from each category: ", result.sum())
    
    data_filtered = df.loc[result, ['guid','date', 'headline']].sample(frac=1, random_state=13)
    return data_filtered
```

```python
list_of_filtered_dfs = [filter_keywords(df) for df in list_of_df]
```

```python
for i,df in enumerate(list_of_filtered_dfs):
    df.to_csv(os.path.join('/projectnb/llamagrp/other_countries_news_framing/',
                             os.path.splitext(file_names[i])[0]+'.csv'),header=False,index=False)
```
