```python
import pandas as pd 
from sklearn.model_selection import StratifiedKFold
import numpy as np
```

```python
data = pd.read_csv('german_326-Copy1.csv', header=0)
data.head()
```

```python
k=8
skf = StratifiedKFold(n_splits=k, random_state=50, shuffle=True)
X = data['text_a']
y = data['theme1']
skf.get_n_splits(X,y)
```

```python
for train_index, test_index in skf.split(X,y):
    print("dev:", len(train_index), "few shot train:", len(test_index))
    dev, few_shot_train = data.loc[train_index], data.loc[test_index]
    break
```

```python
dev_326 = pd.read_csv('dev-Copy1.tsv', sep='\t', header=None)
dev_326.loc[dev.index]
```
