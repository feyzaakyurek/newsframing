```python
import pandas as pd
import numpy as np
import math
```

```python
headline = pd.read_csv('englishAnnotatedW2WToGerman.txt', sep="\t", header=None)
theme1 = pd.read_csv('englishFirstTheme.txt', sep="\t", header=None)
theme2 = pd.read_csv('englishSecondTheme.txt', sep="\t", header=None)

df = pd.DataFrame({"text_a" : headline[0],
                  "text_b" : pd.Series("a", index=np.arange(len(headline))),
                  "theme1" : theme1[0],
                  "theme2" : theme2[0]})
df
```

```python
for i,frame in enumerate(df['theme2']):
    if math.isnan(frame) or frame == 99:
        df.iat[i,3] = df.iat[i,2]

display(df.head())
```

```python
size = len(df)
labels = np.zeros((size, 9))
labels[np.arange(size), df['theme1']-1] = 1
labels[np.arange(size), (df['theme2']-1).astype(int)] = 1

display(labels[:5,:])
```

```python
labels_df = pd.DataFrame(labels)
labels_df.columns = np.arange(9)+1
labels_df.head()
```

```python
df_wide = pd.DataFrame({"text_a": df['text_a'],
                        "text_b": df['text_b']})
df_wide = pd.concat([df_wide, labels_df], axis = 1)
display(df_wide)
```

```python
df_wide.to_csv('train.tsv', sep='\t', index=True, header=False)
```
