```python
import pandas as pd
```

```python
train = pd.read_csv('train-Copy1.tsv', sep='\t', header=None)
train.head()
```

```python
english_words = []
german_words = []

with open('english_words.txt', 'r') as eng:
    for w in eng:
        english_words.append(w.strip())
        
with open('german_words.txt', 'r') as tr:
    for w in tr:
        german_words.append(w.strip())
print(len(english_words))
print(len(german_words))
```

```python
d = {}
for i,e in enumerate(english_words):
    d[e] = german_words[i]

d['US'] = 'UNS.'
d['U.S.'] = 'UNS.'
d['Trump'] = 'Trump'
d['Twitter'] = 'Twitter'
d
```

```python
def replace_words(words, d):
    new_words = []
    for w in words:
        if w in d.keys():
            new_words.append(d[w])
            print(w, d[w])
        elif w.lower() in d.keys():
            new_words.append(d[w.lower()])
            print(w, d[w.lower()])
        else:
            new_words.append(w)
    return ' '.join(new_words)
```

```python
code_switched = []
for line in train[1]:
    words = line.split(' ')
    code_switched.append(replace_words(words, d))
```

```python
code_switched
```

```python
train[1] = code_switched
train
```

```python
train.to_csv('train.tsv', sep='\t', header=False, index=False)
```
