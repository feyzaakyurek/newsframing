```python
from sklearn.model_selection import train_test_split, StratifiedKFold
import pandas as pd
import numpy as np
# import tensorflow as tf
# import tensorflow_hub as hub
# from datetime import datetime
```

```python
#read source data from csv file
english = pd.read_csv('dataset/english.csv')
display(english.columns)
print("Length of data: ", len(english))

english = pd.DataFrame({"ImageID": english['ImageID'],
                        "news_title": english['news_title'],
                        "theme1": english['Q3 Theme1'],
                        "theme2": english['Q3 Theme2']})
display(english.head())
size = len(english)
#create test and train split
# train, test_labeled = train_test_split(all_data_english, test_size=0.2, random_state = 45)
```

```python
# alphas=[0.248, 0.044, 0.025, 0.145, 0.069, 0.083, 0.040, 0.230, 0.118]
```

```python
#create stratified kfolds, k = 5
k = 5
skf = StratifiedKFold(n_splits=k, random_state=50, shuffle=True)
X = english['news_title']
y = english['theme1'] 
skf.get_n_splits(X, y)
print(skf)
```

```python
#delete flags from tensorflow
# del_all_flags(tf.flags.FLAGS)

# !!!delete bert_output/checkpoint file if exists.

for idx, (train_index, test_index) in enumerate(skf.split(X, y)):
    print("Fold: ", idx, "TRAIN:", len(train_index), "TEST:", len(test_index))
    train, test = english.loc[train_index], english.loc[test_index]
    display(train.head())
    display(test.head())
    
    path = 'dataset/dataset_multiclass/'+str(idx)+'/'
    train.to_csv(path+'train.tsv', sep='\t', index=False, header=False)
    test.to_csv(path+'test_labeled.tsv', sep = '\t', index=False, header=False)
    test = pd.DataFrame({'ImageID': test['ImageID'], 'news_title': test['news_title']})
    test.to_csv(path+'test.tsv', sep='\t', index=False, header=True)
                         
```

```python
import pandas as pd
import numpy as np
```

```python
#POST-PROCESS

accuracies = []
k = 5
for fold in range(k):
    path_dataset = 'dataset/'+str(fold)+'/'
    df_test = pd.read_csv(path_dataset+'test_labeled.tsv', sep = '\t')
    display(df_test.head())
    path_output = 'bert_output/'+str(fold)+'/'
    df_result = pd.read_csv(path_output+'test_results.tsv', sep = '\t', header = None)

    df_map_result = pd.DataFrame({'text':df_test['text_a'], 
                              'gold_label':df_test['label'],
                              'predicted_label':df_result.idxmax(axis=1)+1})
    acc = np.sum(np.equal(df_map_result['gold_label'], df_map_result['predicted_label']))/len(df_result)
    accuracies.append(acc)
    print("SAMPLES FROM PREDICTION, FOLD: ", fold)
    print(df_map_result.sample(4, random_state=10))
    print("Accuracy:%.2f\nTest Size:%d" % (acc, len(df_result)))
    
print("All accuracies: ", accuracies)
print("Average accuracy for ", str(k)," folds: ", np.mean(accuracies))
```
